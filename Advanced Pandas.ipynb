## <center> Working with dates

```python
import pandas as pd
```

###Generate series of dates with `period_range`

```python
daterange = pd.period_range('1/1/2020', freq='30d', periods=4)
```

```python
date_df = pd.DataFrame(data=daterange,columns=['sample date'])
date_df
```

### Date difference from prior date using `diff`

```python
date_df['date difference'] = date_df['sample date'].diff(periods=1)
date_df
```

### Find the first day of the month

```python
date_df['first of month'] = date_df['sample date'].values.astype('datetime64[M]')
date_df
```

### Date types

```python
date_df.dtypes
```

```python
date_df['sample date'] = date_df['sample date'].dt.to_timestamp()
date_df.dtypes
```

### Date Subtraction

```python
date_df['sample date'] - date_df['first of month']
```

```python
date_df['sample date'] - date_df['date difference']
```

```python
date_df['sample date'] - pd.Timedelta('30 d')
```

### More datetime properties with `dt`

```python
date_df['sample date'].dt.day_name()
```

## <center> `Apply`, `Map`, and `Applymap` in Pandas

```python
df = pd.DataFrame({"Region":['North','West','East','South','North','West','East','South'],
          "Team":['One','One','One','One','Two','Two','Two','Two'],
          "Squad":['A','B','C','D','E','F','G','H'],
          "Revenue":[7500,5500,2750,6400,2300,3750,1900,575],
            "Cost":[5200,5100,4400,5300,1250,1300,2100,50]})
```

## Use `apply()` to alter values along an axis in your **dataframe** or in a **series** by applying a function. lambda function: to apply a function wothout needing to create it in advance

```python
df['Profit'] = df.apply(lambda x: 'Profit' if x['Revenue']>x['Cost'] else 'Loss',axis=1)
df
```

## Use `map` to substitute each value in a **series**, using either a function, dictionary, or series.

```python
team_map = {"One":"Red","Two":"Blue"}
```

```python
df['Team Color'] = df['Team'].map(team_map)
df
```

## Use `applymap()` to apply a function to each element in your **dataframe**

```python
df.applymap(lambda x: len(str(x)))
```

## If all else fails, use a `for` loop.

```python
new_col = []

for i in range(0,len(df)):
    rev = df['Revenue'][i]/df[df['Region']==df.loc[i,'Region']]['Revenue'].sum()
    new_col.append(rev)
```

```python
df['Revenue Share of Region'] = new_col
df.sort_values(by='Region')
```

Allows you to take a variable separating your rows and *pivot* that to your columns.

<b> Note:</b> does not support aggregation and therefore requires a unique index.

```python
df = pd.DataFrame({"Region":['North','West','East','South','North','West','East','South'],
          "Team":['One','One','One','One','Two','Two','Two','Two'],
          "Revenue":[7500,5500,2750,6400,2300,3750,1900,575],
            "Cost":[5200,5100,4400,5300,1250,1300,2100,50]})
df
```

```python
df.pivot(index='Region',columns='Team',values='Revenue')
```

<img src="https://pandas.pydata.org/pandas-docs/stable/_images/reshaping_stack.png">

Pivot a level of column labels to rows. Work with a multiindex.

```python
df2 = df.set_index(['Region','Team'])
```

```python
stacked = pd.DataFrame(df2.stack())
stacked
```

<img src = "https://pandas.pydata.org/pandas-docs/stable/_images/reshaping_unstack.png">

Opposite of `stack` - pivots level of row labels to columns.

```python
stacked.unstack('Region')
```

<img src="https://pandas.pydata.org/pandas-docs/stable/_images/reshaping_melt.png">

```python
df.head(3)
```

```python
df.melt(id_vars=['Region','Team'], var_name='value type')
```

## <center> Merging DataFrames

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'letter': ['A', 'B', 'C', 'D'],
                    'number': [1, 2, 3, 4]})
df2 = pd.DataFrame({'letter': ['C', 'D', 'E', 'F'],
                    'number': [3, 4, 5, 6]})
```

### <b> Left Join
<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/f/f6/SQL_Join_-_01_A_Left_Join_B.svg/330px-SQL_Join_-_01_A_Left_Join_B.svg.png">

```python
df1.merge(df2,how='left',on='number')
```

### <b> Inner join
<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/1/18/SQL_Join_-_07_A_Inner_Join_B.svg/330px-SQL_Join_-_07_A_Inner_Join_B.svg.png">

```python
df1.merge(df2,how='inner',left_on='number',right_on='number')
```

### <b> Right Join
<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/5/5f/SQL_Join_-_03_A_Right_Join_B.svg/330px-SQL_Join_-_03_A_Right_Join_B.svg.png">

```python
df1.merge(df2,how='right',on='number',suffixes=('','_right'))
```

### Union with `pd.concat`

```python
# drop duplicates with .drop_duplicates()
df3 = pd.concat([df1,df2]).drop_duplicates().reset_index(drop=True)
df3
```

### Concatenate dataframes horizontally

```python
df4 = pd.concat([df1,df2],axis=1)
df4
```

### Append new row to your dataframe

```python
new_row = pd.Series(['Z',26],index=df3.columns)
df3.append(new_row,ignore_index=True)
```

### Join along your index

```python
join_df = pd.DataFrame({'letter': ['F','G', 'H', 'I'],
                        'number': [6, 7, 8, 9]})
```

```python
df2.join(join_df, rsuffix='_right')
```

## <center> Categorizing and Labelling Data

```python
import pandas as pd
```

```python
df = pd.DataFrame({"Species":['Chinook','Chum','Coho','Steelhead','Bull Trout'],
             "Population":['Skokomish','Lower Skokomish','Skokomish','Skokomish','SF Skokomish'],
             "Count":[1208,2396,3220,6245,8216]})
df
```

### Binning Numerical Data with `pd.cut`

```python
import numpy as np
bins = [0, 2000, 4000, 6000, 8000, np.inf]
labels = ['Low Return', 'Below Avg Return', 'Avg Return', 'Above Avg Return', 'High Return']
```

```python
df['Count Category'] = pd.cut(df['Count'], bins, labels=labels)
df
```

### Map Species to Endangered Status

```python
fed_status ={"Chinook":"Threatened",
"Chum":"Not Warranted",
"Coho":"Not Warranted",
"Steelhead":"Threatened"}
```

```python
df['Federal Status'] = df['Species'].map(fed_status)
df
```

### Introducing the `categorical` data type

```python
df['Count Category'] = pd.Categorical(df['Count Category'],
               ordered=True,
               categories=labels)
df['Count Category']
```

```python
df.sort_values(by=['Count Category'],ascending=False)
```

### Use `get_dummies()` to Convert a Categorical Variable into a Dummy Variable

```python
pd.get_dummies(df['Count Category'])
```
